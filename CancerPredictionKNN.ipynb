```python
import numpy as np
import csv



def get_distance(list_a, list_b):
    # converting list to array
    arr_a = np.array(list_a)
    arr_b = np.array(list_b)

    return np.linalg.norm(arr_a - arr_b)

#print(get_distance(list_a, list_b))


def load_from_csv(file_path):
    matrix = []
    with open(file_path, 'r') as file:
        reader = csv.reader(file, quoting=csv.QUOTE_NONNUMERIC)
        for row in reader:
            matrix.append(row)
           
        return matrix

#print(load_from_csv('Learning_Data.csv'))

#M = [[12,13,15,17,19], [25,32,16,8,9], [3,2,4,16,18], [15,19,16,7,6], [5,3,11,14,17]]

def get_standard_deviation(matrix, column):
   
    sum_of_column = 0
    sum_of_squares = 0
    num_rows = len(matrix)
   
    for i in range(num_rows):
        sum_of_column+=matrix[i][column]
       
    column_avg = sum_of_column/num_rows
   
    for i in range(num_rows):
        sum_of_squares+=(matrix[i][column]-column_avg)**2
   
    standard_deviation = (sum_of_squares/(num_rows-1))**0.5
    return standard_deviation

#get_standard_deviation(M,4)

def get_standardised_matrix(matrix):
   
    avg_array = []
    std_array = []
    num_rows = len(matrix)
    num_columns = len(matrix[0])
    sum_column = 0
   
    #print(num_rows, num_columns)
    for i in range(num_columns):
        sum_column = 0
        for j in range(num_rows):
            sum_column+=matrix[j][i]
       
        column_avg = sum_column/num_rows
        avg_array.append(column_avg)
        std_array.append(get_standard_deviation(matrix, i))
   
    #print(avg_array)
    #print(std_array)
   
    std_row = []
    std_matrix = []
   
    for i in range(num_rows):
        for j in range(num_columns):
            std_value = (matrix[i][j]-avg_array[j])/std_array[j]
            std_row.append(std_value)
        std_matrix.append(std_row)
        std_row = []
   
    return std_matrix
 

#load_from_csv('Data.csv')


#get_standard_deviation(load_from_csv('Data.csv'), 8)
#load_from_csv('Data.csv')
#get_standardised_matrix(load_from_csv('Data.csv'))


ref_row = [0.18401406215872845, -0.04773091604150719, -0.39804070690259596, 1.9210909925986457, 0.8117593582997904, 1.6270264378870563, 2.2104714898633753, -0.6281680148633517, 0.234118431659613]

def get_k_nearest_labels(ref_row, std_learning_data_matrix, labels_matrix, k):
   
    distance_dict = {}
   
    for i in range(len(std_learning_data_matrix)):
        #print(data_matrix[i])
        distance_dict[i]=get_distance(ref_row, std_learning_data_matrix[i])
        #print(get_distance(ref_row, data_matrix[i]))
   
    sorted_distance = sorted(distance_dict.items(), key=lambda x: x[1])
    #print(sorted_distance)
   
   
    k_nearest = sorted_distance[:k]
    label_matrix = []
   
    for i in range(k):
        label_matrix.append(labels_matrix[k_nearest[i][0]])
       
    return label_matrix  
   
def get_mode(label_matrix):
    labels = []
    for i in range(len(label_matrix)):
         labels.append(label_matrix[i][0])
    mode = max(set(labels), key=labels.count)
    return mode

#print(labels_matrix)    

#get_mode(get_k_nearest_labels(ref_row, data_matrix, labels_matrix, 20))

data_matrix = load_from_csv('Data.csv');
learning_data_matrix = load_from_csv('Learning_Data.csv');
labels_matrix = load_from_csv('Learning_Data_Labels.csv');
correct_data_labels = load_from_csv('Correct_Data_Labels.csv');

def classify(data_matrix, learning_data_matrix, labels_matrix, k):
   
    std_data_matrix = get_standardised_matrix(data_matrix)
    std_learning_data_matrix = get_standardised_matrix(learning_data_matrix)
    data_labels_matrix = []
   
    for i in range(len(std_data_matrix)):
        data_labels_matrix.append([get_mode(get_k_nearest_labels(std_data_matrix[i], std_learning_data_matrix, labels_matrix, k))])
   
    return data_labels_matrix
   



#print(classified_data_labels)
#print(correct_data_labels)

def get_accuracy(correct_data_labels, classified_data_labels):
    if len(correct_data_labels)==len(classified_data_labels):
        score = 0
        for i in range(len(correct_data_labels)):
            if correct_data_labels[i]==classified_data_labels[i]:
                score=score+1
        accuracy = score/len(correct_data_labels)
        return accuracy
    else:
        print('Data not same')

       
def run_test():
    data_matrix = load_from_csv('Data.csv');
    learning_data_matrix = load_from_csv('Learning_Data.csv');
    labels_matrix = load_from_csv('Learning_Data_Labels.csv');
    correct_data_labels = load_from_csv('Correct_Data_Labels.csv');
   
    for i in range(3,21):
        classified_data_labels = classify(data_matrix, learning_data_matrix, labels_matrix, i)
        accuracy = round(get_accuracy(correct_data_labels, classified_data_labels), 2)
        print('k='+str(i)+', Accuracy = '+str(accuracy*100)+'%')
   
   
run_test() 
```
